## Assignment 1 :Create a list of sandwich ingredients that you can think of. Use slicing to print some of the ingredients

```python
# function to display User option menu and return the option selected by the user
def ask_user_input():
  print('\nChoose the list operations from below:\n 1:Print the list\n 2:Print the total number of items on the list',
  '\n 3:Print the first 3 ingredients',
  '\n 4:Print the last 3 ingredients',
  '\n 5:Search for an ingredient'
  '\n 6:add an ingredient to the list',
  '\n #:Exit Menu'    )
  return(input())
```

```python
#function to append the list
def add_ing(ing):
  sw_ingredients.append(ing)
```

```python
#function to search an item in the list
def search_ing(ing):
  try:
    print('The item is found at position ',sw_ingredients.index(ing)+1)
  except:
    print(ing,' does not exist in the list')
```

```python
# Let the user build a list of sandwich ingredients
print('enter the sandwich ingredients that you can think of, separated by comma')
# Use this list - Bread,Cheese,Salami,Spread,Lettuce,Tomatoe,Avocado,Eggs,Ketchup
sw_input = input()
sw_ingredients = sw_input.split(',')
```

```python
#Perform actions chosen by the user on sandwich ingredients list
menu_sel= ask_user_input()
while menu_sel != '#':
  if menu_sel=='1':
    print(sw_ingredients)
  elif menu_sel=='2':
    print('The total number of items on the list is:',len(sw_ingredients))
  elif(menu_sel=='3'):
    print('The first three ingredients are:')
    print(sw_ingredients[0:3])
  elif(menu_sel=='4'):
    print('The last three ingredients are:')
    print(sw_ingredients[len(sw_ingredients)-3:len(sw_ingredients)])
  elif(menu_sel=='5'):
    search_ing(input('Please enter the item you want to search in the list:'))
  elif(menu_sel=='6'):
    new_item=input('Please enter the ingredient you want to add to the list')
    add_ing(new_item)
    print('Your new list of ingredients is:',sw_ingredients)
  elif(menu_sel=='#'):
    break
  else:
    print('Incorrect option chosen')
  menu_sel= ask_user_input()
```
