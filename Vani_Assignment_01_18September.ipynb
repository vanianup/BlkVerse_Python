Assignment1:Using object oriented programming, create a book log for books read by the user .
Below program allows the user to create a book log. User can record the details of the book they read and the rating they assign to the book. They can print the log and search for a particular book title.

```python
my_book_log=[]
class book_read:
    #book_read is a class defining the basic attributes of a book
    def __init__(self,title,author,genre,pages):
        self.book_title = title
        self.book_author = author
        self.book_genre = genre
        self.book_size = pages


class book_log_entry:
  #book_log_entry is a class for logging in the details of a book that was read 
    def __init__(self,start_date,end_date,rating,book):
        self.start_date=start_date
        self.end_date=end_date
        self.rating=rating
        self.book=book
        
    # define a method to add the log entries to a global list   
    def add_book(self,log_entry):
        log_entry.append(self)


```

```python
def build_menu():
#build a user menu for basic operations on the book log
    print('                              Book log Menu')
    print('************************************************************************')
    print('1: Add book to your book log')
    print('2: Print the book log with book title and date of reading ')
    print('3: Search a book in the log')
    print('#: Quit')
    option = input('Choose your option for the book log: ')
    return option

def accept_book():
#take user input for a book to be added to the book log
    print('Enter the details of the book you want to add to the log')
    book=book_read(input('Book Title: '),
                           input('Author: '),
                           input('Genre: '),
                           input('Number of pages: '))
    return book
def make_log_entry(book):
# add a book to the read log    
    print('Log the time taken to read and your review rating:')
    log_entry= book_log_entry(input('Enter start date: '),
                             input('Enter end date: '),
                              input('Enter your rating: '),book)
    print('Your entry is added successfully\n')
    return log_entry

def print_book_log():
    print('Below books are added to your reading log\n')
    for i in my_book_log:
        print('Title: ',i.book.book_title)
        print('Started on: ',i.start_date,'\n')

def search_book_log():
    title = input('Enter the title you want to search: ')
    occurance = 0
    for i in my_book_log:
       
        if (i.book.book_title.lower()==title.lower()):
            print(title,' is found in the log with below details:')
            print('Title: ',i.book.book_title)
            print('Author: ',i.book.book_author)
            print('Genre: ',i.book.book_genre)
            print('Rating: ',i.rating)
            occurance = 1
            break
    print(occurance)
    if (occurance==0):
        print(title,' is not in your book log\n')
```

```python
#Main program 

user_input=''

while (user_input !='#'):
    user_input=build_menu()
    if (user_input == '1'):
        new_book_read=accept_book()
        my_log_entry=make_log_entry(new_book_read)
        my_log_entry.add_book(my_book_log)
    elif (user_input == '2'):
        print_book_log()
    elif (user_input == '3'):
        search_book_log()
    elif (user_input == '#'):
        break
    else:
        print('Incorrect menu option entered')
    print('\n')    
    


    

```
